```python
def headers_to_snake_case(df):
    """Renames headers to snake_case"""
    df = df.copy()

    df.columns = (
        df.columns.astype(str)
        .str.strip()
        .str.replace(r"([A-Z]+)([A-Z][a-z])", r"\1_\2", regex=True)
        .str.replace(r"([a-z0-9])([A-Z])", r"\1_\2", regex=True)
        .str.lower()
        .str.replace(r"[^a-z0-9]+", "_", regex=True)
        .str.strip("_")
    )

    return df
```

```python
names_mapper = name_df.set_index('todo_file')["Name"]
rows = []

for file_name, df in todo_dfs.items():
   counts = df[['Emails', 'ToDo']].count()
   rows.append({
        'user': names_mapper.get(file_name),
        'emails_count': counts['Emails'],
       'todo_counts': counts['ToDo']
   }
   )

todo_summary = pd.DataFrame(rows).sort_values(by='user')

todo_summary
```
